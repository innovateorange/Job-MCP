Here's a prompt to extract information from a resume into a structured JSON format:

```
Extract the following information from the provided resume text:

- **Name**: The full name of the applicant.
- **Email**: The primary email address of the applicant.
- **Skills**: A list of technical and soft skills, categorized if possible (e.g., Programming Languages, Tools, Methodologies).
- **Experience**: A list of professional experiences. For each experience, include:
  - Job Title
  - Company Name
  - Location (City, State/Country)
  - Start Date (Month Year, if available)
  - End Date (Month Year, or 'Present' if current)
  - Key Responsibilities/Achievements (a bulleted list)
- **Education**: A list of educational qualifications. For each entry, include:
  - Degree/Qualification
  - Institution Name
  - Location (City, State/Country)
  - Graduation Date (Month Year, or Year, if available)

Output the extracted information as a JSON object with the following structure:

```json
{
  "name": "string",
  "email": "string",
  "skills": {
    "programming_languages": ["string"],
    "tools": ["string"],
    "methodologies": ["string"],
    "other": ["string"]
  },
  "experience": [
    {
      "job_title": "string",
      "company_name": "string",
      "location": "string",
      "start_date": "string",
      "end_date": "string",
      "responsibilities": ["string"]
    }
  ],
  "education": [
    {
      "degree": "string",
      "institution": "string",
      "location": "string",
      "graduation_date": "string"
    }
  ]
}
```

**Resume Text:**
[Insert resume text here]
```

````python
import docx

def get_docx_text(file_path):
    doc = docx.Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

resume_file_path = '/content/Sebastian Sares Resume-  Summer 2025 (1).docx'
resume_text = get_docx_text(resume_file_path)

print("Extracted Resume Text (first 500 characters):\n")
print(resume_text[:500])

# Store the full resume text for later use
# This will be used in the next step to pass to the generative model
# if the user proceeds with the extraction.
# You can verify this variable in the kernel.
resume_text_full = resume_text
````
